What is Redux?

Redux is a predictable state management library for JavaScript applications.

It is commonly used with React, but it can also work with other frameworks or even vanilla JavaScript.

Redux helps manage application-wide state in a centralized and structured way.

#Q3 What is Redux Thunk and how does it facilitate asynchronous logic in Redux*

_
What is Redux Thunk?

Redux Thunk is a middleware for Redux that allows you to write asynchronous logic inside action creators.

By default, Redux only supports synchronous actions ‚Äî meaning you can only dispatch plain JavaScript objects like:

{ type: "FETCH_DATA" }


But real applications often need to:

Fetch data from an API

Wait for a response

Dispatch multiple actions (loading, success, error)

Redux Thunk makes this possible.

#Q4 What are the benefits of using the Redux Toolkit*?
_
What is Redux Toolkit?

Redux Toolkit (RTK) is the official, recommended way to write Redux logic for modern applications using Redux.

It was created to solve common problems like:

Too much boilerplate

Complex setup

Accidental state mutations

Difficult async logic handling

It works seamlessly with React and is now the standard way to use Redux.
